The following notebook contains some of the results of the DDPG algorithm. The order is from **oldest** (top) to **most recent** (bottom).

```python
import pandas as pd 
import matplotlib.pyplot as plt
import json
from pprint import pprint
```

## DDPG-AlphaWorm-HPO (28.06.2020)

One of the very first runs with using HPO after solving the mountain car set up (`models/MountainCar`). 

```python
fig, ax = plt.subplots(figsize=(15,5))

rewards = pd.read_csv("../models/DDPG-AlphaWorm/28-06-DDPG-AlphaWorm-HPO/HPO-Results/best_agent/rewards.csv")

rewards["reward"].plot(ax=ax)
ax.set_ylabel("reward (mean)")
ax.set_xlabel("episodes")
```

Below you find some results that are tracked during training.

```python
episodes = [
    "Episode 0",
    "Episode 1",
    "Episode 2",
    "Episode 3",
    "Episode 4",
    "Episode 5",
    "Episode 6",
    "Episode 7",
    "Episode 8",
    "Episode 9",
    "Episode 10",
]

fig, ax = plt.subplots(figsize=(15,5))

rewards_episode = pd.read_csv("../models/DDPG-AlphaWorm/28-06-DDPG-AlphaWorm-HPO/HPO-Results/best_agent/trainings_rewards.csv")

rewards_episode[episodes].plot(ax = ax)
ax.legend()
ax.set_ylabel("reward (mean)")
ax.set_xlabel("trainings steps")
```

## DDPG-AlphaWorm-7-BaselineTraining (28.06.2020)

This experiment is conducted by the baseline training procedure [https://github.com/openai/baselines/blob/master/baselines/ddpg/ddpg.py](https://github.com/openai/baselines/blob/master/baselines/ddpg/ddpg.py).

```python
fig, ax = plt.subplots(figsize=(15,5))

rewards = pd.read_csv("../models/DDPG-AlphaWorm/28-06-DDPG-AlphaWorm-7-BaselineTraining/rewards.csv")

rewards["reward"].plot(ax=ax)
ax.set_ylabel("reward (mean)")
ax.set_xlabel("nb_epochs")
```

## DDPG-AlphaWorm-8-FixedParameter (29.06.2020)

One of the very first runs without HPO after 
solving the mountain car set up (`models/MountainCar`).

This experiment is conducted by the naive training procedure (`ddpg_trainer.train()`)

```python
fig, ax = plt.subplots(figsize=(15,5))

rewards = pd.read_csv("../models/DDPG-AlphaWorm/29-06-DDPG-AlphaWorm-8-FixedParameter/best_agent/rewards.csv")

with open("../models/DDPG-AlphaWorm/29-06-DDPG-AlphaWorm-8-FixedParameter/best_agent/DPPG-WormDomain-8-FixedParameter_best_params.json") as json_file:
    pprint(json.load(json_file))

rewards["reward"].plot(ax=ax)
ax.set_ylabel("reward (mean)")
ax.set_xlabel("episodes")
```

## DDPG-Pendulum (06.07.2020)

This experiment was conducted by using the ddpg baseline training procedure ([https://github.com/openai/baselines/blob/master/baselines/ddpg/ddpg.py](https://github.com/openai/baselines/blob/master/baselines/ddpg/ddpg.py)) that differs from the naive way. The ddpg baseline procedure has more loops and inner circle. Unfortunately, doesn't lead to a successful training.

```python
fig, ax = plt.subplots(figsize=(15,5))

rewards = pd.read_csv("../models/DDPG-Pendulum/06-07-DDPG-Pendulum/DPPG-Pendulum-v0-2-Longer-Training/eval_rewards.csv")

with open("../models/DDPG-Pendulum/06-07-DDPG-Pendulum/DPPG-Pendulum-v0-2-Longer-Training/config.json") as json_file:
    pprint(json.load(json_file))

rewards["0"].plot(ax=ax)
ax.set_ylabel("reward (mean)")
ax.set_xlabel("episodes")
```

Some remarks:

Unfortunately the DDPG doesn't work very well on the Unity Worm Domain and we could not figure out what the issue was. We did more trainings runs with different paramter set up (e.g. larger networks, different number of episodes, different number of trainings steps, etc.). All results look similar.
